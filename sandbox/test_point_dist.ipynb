```
import numpy as np

def dist(pts):
    num = pts.shape[0]
    for i in range(num):
        x1 = pts[i][0]
        y1 = pts[i][1]
        x2 = pts[(i + 1) % num][0]
        y2 = pts[(i + 1) % num][1]
        dist = pow(pow(x1 - x2, 2) + pow(y1 - y2, 2), 0.5)
        print("|({}, {}) - ({}, {})| = {}".format(x1, y1, x2, y2, dist))
```
