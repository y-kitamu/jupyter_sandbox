```
import numpy as np

def dist(pts):
    num = pts.shape[0]
    for i in range(num):
        x1 = pts[i][0]
        y1 = pts[i][1]
        x2 = pts[(i + 1) % num][0]
        y2 = pts[(i + 1) % num][1]
        dist = pow(pow(x1 - x2, 2) + pow(y1 - y2, 2), 0.5)
        print("|({}, {}) - ({}, {})| = {}".format(x1, y1, x2, y2, dist))
```

```
# 20190701

def calc_obj_pts(rot, trans, intri, offset, ipt, height):
    rot_trans = rot.T * trans
    # print("{}".format(rot_trans.T))
    tmp = rot.T * np.linalg.inv(intri) * ipt
    opt = tmp * (-height + rot_trans[2]) / tmp[2] - rot_trans - offset
    return opt

## SBV013

### pose01
intri = np.matrix([[3336.560020139034, 0, 1512],
                   [0, 3328.484405616405, 2016],
                   [0, 0, 1]])

rot = np.matrix([[0.098056,  0.829827, 0.549338],
                 [-0.975981,  0.188092, -0.10992],
                 [-0.19454, -0.525366, 0.828339]])

trans = np.matrix([[123.928],
                   [129.107],
                   [1612]])

offset = np.matrix([[-1836], [-1483], [0]])

ipt0 = np.matrix([[1013], [3854], [1]])
ipt1 = np.matrix([[1017], [3052], [1]])
opt0 = calc_obj_pts(rot, trans, intri, offset, ipt0, 180)
opt1 = calc_obj_pts(rot, trans, intri, offset, ipt1, 180)

# print("{}".format(opt0.T))
# print("{}".format(opt1.T))

print(np.linalg.norm(opt0 - opt1), 2)

obs_pts = np.array([[1276, 1054], [766, 1098]])
dist(obs_pts)

obs_opt0 = np.matrix([[954], [1244], [-180]])
obs_opt1 = np.matrix([[1402], [1198], [-180]])

print(np.linalg.norm(opt0 - obs_opt0, 2)) 
print(np.linalg.norm(opt1 - obs_opt1, 2)) 


### pose02
intri = np.matrix([[3336.560020139034, 0, 1512],
                   [0, 3328.484405616405, 2016],
                   [0, 0, 1]])
rot = np.matrix([[0.833096, -0.0606507,   0.549794],
                 [0.143636,   0.983593,  -0.109145],
                 [-0.534154,   0.169898,   0.828139]])
trans = np.matrix([[-181.938], [-832.818], [1685]])
offset = np.matrix([[0], [0], [0]])

ipt0 = np.matrix([[1025], [3045], [1]])
ipt1 = np.matrix([[1013], [3851], [1]])

opt0 = calc_obj_pts(rot, trans, intri, offset, ipt0, 0)
opt1 = calc_obj_pts(rot, trans, intri, offset, ipt1, 0)

print("theoretical val = {}".format(np.linalg.norm(opt0 - opt1, 2)))

obs_pts = np.array([[1039, 1786], [1062, 2297]])
dist(obs_pts)

```
